```python
import torch.nn as nn
```

```python
torch.manual_seed(1337)
B, T, C = 4, 8, 32
x = torch.randn((4, 8, 32))

head_size = 16
key = nn.Linear(C, head_size, bias=False) #(input_dim = C, output_dim = head_size)
query = nn.Linear(C, head_size, bias=False) #(input_dim=C, output_dim=head_size)
value = nn.linear(C, head_size, bias=False)


k = key(x) # (B, T, head_size)
q = query(x) # (B, T, head_size)
v = value(x) # (B, T, head_size)

#attention value = query @ key(matmul) 
attention_value = q @ k.transpose(-2, -1) # (B, T, head_size) * (B, head_size, T) = (B, T, T)
wei = attention_value
# tril is aiming to only keep the lower part 

tril = torch.tril(torch.ones(T, T))
wei = wei.masked_fill(tril == 0, float('-inf')) # (B, T, T)--> match
wei = F.softmax(wei, dim=-1) #shape (B, T, T)
out = wei @ v # (B, T, T) * (B, T, head_size) = (B, T, head_size)
out.shape
```
